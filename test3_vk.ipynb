### Аналитическое решение

- Решим задачу, найдя вероятность противоположного события: игрок не получит полный сет после успешного прохождения ивента.

- По определению вероятности, если $A$ -- событие "**получение игроком полного сета после успешного прохождения ивента**" и $\bar A$ -- событие "**неполучение игроком полного сета после успешного прохождения ивента**", то тогда $1 - P(\bar A) = P(A)$, где $P(A)$ и $P(\bar A)$ -- вероятности указанных выше событий. Найдя вероятность события $\bar A$, мы сможем легко получить вероятность события $A$.

- Игрок **не получит полный сет**, если в его сундуках выпадет **только два типа предметов**, причем неважно, в каком соотношении, то есть неудачные комбинации предметов это такие, где есть всего $2$ типа предметов. Так как всего существует $3$ типа предметов, то у нас есть **3 основных случая**: первый, когда в сундуках есть только предметы $1$ и $2$ типов, второй, когда в сундуках есть предметы только $1$ и $3$ типов, и третий, когда в сундуках есть предметы только $2$ и $3$ типов.

- Далее будем рассуждать так. Наши неудачные комбинации можем представить как **бинарные векторы**, неважно, что обозначать за $0$, а что за $1$ (то есть не имеет значения, обозначим мы за $0$ предмет первого типа или предмет второго типа в комбинации из предметов только первого и второго типов). 

- Далее посчитаем возможное число таких комбинаций для каждого из трех случаев: 
    - у нас могут быть все предметы первого типа, может быть пять предметов первого типа и один предмет второго типа и т.д.;
    - так как существует $6$ сундуков, то, например, пять предметов первого типа могут быть в первых пяти сундуках, а один предмет второго типа -- в шестом. а может быть и наоборот (второго типа -- в первом, первого -- в остальных)
    - получаем сумму числа сочетаний $\sum C_6^k$, где $k$ -- будет меняться от $0$ до $6$;
    - из комбинаторики известно, что данная сумма равна $2^6$;
    - и так будет для каждого из трех основных случаев, то есть всего получаем $3 \cdot 2^6$ неудачных случаев.
    
- Далее посчитаем сколько возможно случаев всего (как удачных, так и неудачных). Это число равно $3^6$, поскольку предметы в сундуках выпадают независимо, то для каждого сундука всегда существует $3$ возможных предмета.

- Вероятность неудачи, таким образом, равна:
$$
P(\bar A) = \frac{3 \cdot 2^6}{3^6} = \frac{2^6}{3^5} \approx 0.2634.
$$

- Вероятность успеха:
$$
P(A) = 1 - P(\bar A) = 1 - \frac{2^6}{3^5} \approx 0.7366.
$$

- **Ответ: 0.7366**

```python
(2 ** 6) / (3 ** 5) # расчеты вероятности
```

```python
1 - (2 ** 6) / (3 ** 5) # расчеты вероятности
```

### Решение методом Монте-Карло

Для численного решения данной задачи воспользуемся методом Монте-Карло. Данный метод позволит избежать аналитических выкладок, приведенных выше, но при этом потеряет в точности. Будем генерировать выпадение предметов в сундуках с помощью функции `random.choice`, далее если сгенерированное множество имеет мощность, равную $3$ (есть три различных предмета), значит, успех, иначе -- неудача.

```python
import numpy as np
```

```python
N, luck = 100000, 0
for i in range(N):
    if len(set(np.random.choice(3, 6))) == 3: luck += 1
print(f'Решение, полученное методом Монте-Карло {luck / N}')
```

Как видим, решение, полученное численно, совпадает до определенного количества знаков после запятой с решением, полученным аналитически.
